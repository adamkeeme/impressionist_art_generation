```python
import os
import random
from PIL import Image
import matplotlib.pyplot as plt

input_dir = "/Users/pc/Documents/GitHub/impressionist_art_generation/data/wikiart/Impressionism"

files = [
    f for f in os.listdir(input_dir)
    if f.lower().endswith((".png", ".jpg", ".jpeg"))
]

# random pick a file to examine
fname = random.choice(files)
img = Image.open(os.path.join(input_dir, fname)).convert("RGB")

# resize to 512x512 using LANCZOS high quality image filter
resized = img.resize((512, 512), Image.LANCZOS)

plt.figure(figsize=(6,6))
plt.imshow(resized)
plt.title(f"random image: {fname}")
plt.axis("off")
plt.show()
```
